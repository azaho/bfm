## 2. Defining the modules
### Distribution Inverter

```python
class DistributionInverter(BFModule):
    def __init__(self, samples, eps=1e-5):
        super().__init__()
        self.samples = samples
        self.mean = nn.Parameter(samples.mean(dim=0).view(1, 1, -1))
        self.std = nn.Parameter(samples.std(dim=0).view(1, 1, -1))
        self.eps = eps

    def forward(self, x):
        # x is of shape (batch_size, seq_len, n_channels)
        # samples is of shape (n_samples, n_channels)
        return (x - self.mean) / (self.std + self.eps)
        # return torch.sum(self.samples.unsqueeze(0).unsqueeze(0) <= x.unsqueeze(-1), dim=-1) / len(self.samples)
```

### Masker

```python
class Masker(BFModule):
    def __init__(self, n_masked_intervals, interval_length):
        super().__init__()
        self.n_masked_intervals = n_masked_intervals
        self.interval_length = interval_length
        
    def forward(self, x):
        # x shape: (batch_size, seq_len, channels)
        batch_size, seq_len, channels = x.shape
        
        # Initialize mask tensor with all True values
        mask = torch.ones_like(x, dtype=torch.bool)
        
        # For each sequence in the batch
        for b in range(batch_size):
            # Get n_masked_intervals random start positions
            # Make sure intervals don't go beyond sequence length
            max_start = seq_len - self.interval_length
            starts = torch.randint(0, max_start + 1, (self.n_masked_intervals,))
            
            # Create masks for each interval
            for start in starts:
                end = start + self.interval_length
                mask[b, start:end] = False
        
        # Apply mask to input
        masked_x = x * mask
        
        return masked_x, mask
```

### Embedder

```python
class Embedder(BFModule):
    def __init__(self, d_model, resolution=10, range=(-3, 3)):
        super().__init__()
        self.d_model = d_model
        self.resolution = resolution
        self.range = range
        self.delta = (range[1] - range[0]) / (resolution-1)

        self.embedding_centers = torch.linspace(range[0], range[1], resolution)
        self.embedding_weights = torch.nn.Parameter(torch.randn(resolution, d_model) / d_model)
        
    def forward(self, x):
        # x is of shape (batch_size, seq_len, n_channels)
        # output is of shape (batch_size, seq_len, n_channels, d_model)
        output = torch.zeros(*x.shape, self.d_model, device=x.device, dtype=x.dtype)
        for center_i, embedding_center in enumerate(self.embedding_centers):
            distance_multiplier = torch.maximum(1 - torch.abs(x - embedding_center) / self.delta, torch.tensor(0))
            output += distance_multiplier.unsqueeze(-1) * self.embedding_weights[center_i].view(1, 1, 1, -1)
        output[x<self.range[0]] = self.embedding_weights[0]
        output[x>self.range[1]] = self.embedding_weights[-1]
        return output
```

```python
class Model(BFModule):
    def __init__():
        pass

    def generate_continuation(self, batch, n_steps, vocab):
        """
        Generate a continuation of the input sequence for n_steps steps.
        
        Args:
            batch: Input sequence of shape (batch_size, seq_len, 1)
            n_steps: Number of steps to generate
            vocab: Tensor of possible samples to choose from, shape (n_samples, d_model)
        
        Returns:
            Tensor of shape (batch_size, seq_len + n_steps, 1) containing the original sequence
            concatenated with the generated continuation
        """
        # Start with the input batch
        current_sequence = batch.clone()
        # Generate n_steps new timesteps
        for _ in range(n_steps):
            # Get next step prediction
            next_values = self.generate_next_step(current_sequence, vocab).unsqueeze(1)
            # Concatenate with current sequence
            current_sequence = torch.cat([current_sequence, next_values], dim=1)
        return current_sequence
    
class ContrastiveModel(Model):
    def __init__(self, d_model=128, n_layers=4, n_heads=8, d_input=1):
        super(Model, self).__init__()
        self.d_input = d_input
        self.d_model = d_model
        self.n_layers = n_layers
        self.n_heads = n_heads
        self.transformer = Transformer(d_input=d_model, d_model=d_model, d_output=d_model, 
                                        n_layer=n_layers, n_head=n_heads, causal=True, 
                                        rope=True, cls_token=False, rope_base=window_size*2 // d_input)
        self.embed = Embedder(d_model=d_model, resolution=2, range=(-3, 3))
        self.unembed = Embedder(d_model=d_model, resolution=2, range=(-3, 3))
    
    def forward(self, x, y=None):
        # x is of shape (batch_size, seq_len, d_input)
        # y is of shape (batch_size, seq_len, d_input)
        x = self.embed(x).squeeze(-2) # shape (batch_size, seq_len, d_model)
        x = self.transformer(x) # shape (batch_size, seq_len, d_model)

        if y is not None:
            y = self.unembed(y).squeeze(-2) # shape (batch_size, seq_len, d_model)
            return x, y
        return x
        
    def calculate_loss(self, batch):
        # batch is of shape (batch_size, seq_len, d_input)
        x, y = self(batch[:, :-1], batch[:, 1:])
        batch_size = batch.shape[0]

        similarity = torch.matmul(x[:, :].permute(1, 0, 2), y[:, :].permute(1, 2, 0))
        expanded_arange = torch.arange(batch_size).unsqueeze(0).repeat(x.shape[1], 1).to(device, dtype=torch.long).reshape(-1)
        loss = torch.nn.functional.cross_entropy(similarity.view(-1, batch_size), expanded_arange)
        return loss
```

## 3. Evaluation class

```python
import torch
import torch.nn as nn
import torch.optim as optim
import sklearn
from sklearn.metrics import roc_auc_score, accuracy_score
from torch.utils.data import DataLoader
from btbench.btbench_train_test_splits import generate_splits_SS_SM
import numpy as np
import copy

# Assuming n_samples_per_bin and device are defined in the global scope or passed appropriately
# For example:
# n_samples_per_bin = 1 # Define this based on your setup
# device = torch.device("cuda" if torch.cuda.is_available() else "cpu") # Define this based on your setup

class ModelEvaluationFinetune():
    def __init__(self, model, inverter, subject_trials, eval_names, batch_size=128, dtype=torch.float32, 
                 feature_aggregation_method='mean', mean_collapse_factor=1, 
                 lr=1e-4, num_epochs=5, weight_decay=0.01):
        # Store base models to avoid modifying the original ones during fine-tuning across different tasks/splits
        self.base_model = model 
        self.base_inverter = inverter
        self.subject_trials = subject_trials
        self.eval_names = eval_names
        self.batch_size = batch_size
        self.dtype = dtype
        self.feature_aggregation_method = feature_aggregation_method
        self.mean_collapse_factor = mean_collapse_factor
        self.lr = lr
        self.num_epochs = num_epochs
        self.weight_decay = weight_decay
        
        # Create evaluation datasets (pre-loading splits)
        self.evaluation_datasets = {}
        print("Generating evaluation splits...")
        for eval_name in self.eval_names:
            for subject, trial_id in self.subject_trials:
                # Assuming generate_splits_SS_SM returns lists of train/test datasets for cross-validation
                splits = generate_splits_SS_SM(subject, trial_id, eval_name, dtype=self.dtype,
                                               start_neural_data_before_word_onset=0,
                                               end_neural_data_after_word_onset=2048) # Adjust window as needed
                self.evaluation_datasets[(eval_name, subject.subject_identifier, trial_id)] = splits
        print("Finished generating splits.")

    def _get_feature_dim(self, model_output_dim, seq_len_after_collapse):
        if self.feature_aggregation_method == 'mean':
            return model_output_dim
        elif self.feature_aggregation_method == 'concat':
            return model_output_dim * seq_len_after_collapse
        else:
            raise ValueError(f"Unknown feature_aggregation_method: {self.feature_aggregation_method}")

    def _aggregate_features(self, x_embed):
        """Aggregates features based on the specified method."""
        batch_size, seq_len, d_model = x_embed.shape
        
        # Apply mean collapse factor if needed (e.g., average over short time windows first)
        if self.mean_collapse_factor > 1:
             # Ensure seq_len is divisible by mean_collapse_factor
            if seq_len % self.mean_collapse_factor != 0:
                 # Simple truncation if not divisible. Consider padding or error handling.
                 print(f"Warning: seq_len {seq_len} not divisible by mean_collapse_factor {self.mean_collapse_factor}. Truncating.")
                 target_len = (seq_len // self.mean_collapse_factor) * self.mean_collapse_factor
                 x_embed = x_embed[:, :target_len, :]
                 seq_len = target_len

            x_embed = x_embed.reshape(batch_size, seq_len // self.mean_collapse_factor, self.mean_collapse_factor, d_model)
            x_embed = x_embed.mean(dim=2) # Shape: (batch_size, seq_len/mean_collapse_factor, d_model)
        
        # Aggregate across the sequence dimension
        if self.feature_aggregation_method == 'mean':
            features = x_embed.mean(dim=1)  # Average across sequence length -> (batch_size, d_model)
        elif self.feature_aggregation_method == 'concat':
            features = x_embed.reshape(batch_size, -1)  # Concatenate across sequence length -> (batch_size, seq_len*d_model)
        else:
             raise ValueError(f"Unknown feature_aggregation_method: {self.feature_aggregation_method}")
        return features


    def evaluate_on_dataset(self, train_dataset, test_dataset):
        # --- Setup for Fine-tuning ---
        # Create deep copies of the models for this specific fine-tuning task/split
        # This prevents interference between different fine-tuning runs
        current_model = copy.deepcopy(self.base_model).to(device)
        current_inverter = copy.deepcopy(self.base_inverter).to(device)
        
        train_dataloader = DataLoader(train_dataset, batch_size=self.batch_size, shuffle=True, drop_last=True)
        test_dataloader = DataLoader(test_dataset, batch_size=self.batch_size, shuffle=False, drop_last=True)

        # Determine number of classes and feature dimension dynamically
        num_classes = len(np.unique(train_dataset.labels)) # Assumes dataset has .labels attribute
        # Get a sample batch to determine dimensions after model processing
        sample_input, _ = next(iter(train_dataloader))
        sample_input = sample_input.to(device, dtype=self.dtype)
        window_size = sample_input.shape[2]
        # Use global n_samples_per_bin - ensure it's defined correctly in the notebook scope
        sample_data = sample_input.reshape(self.batch_size, window_size // n_samples_per_bin, n_samples_per_bin)
        
        with torch.no_grad(): # Just to get shape, no training here
            sample_data = current_inverter(sample_data)
            sample_embed = current_model(sample_data)
            # Apply potential mean collapse to determine final sequence length for aggregation
            bs, sl, dm = sample_embed.shape
            seq_len_after_collapse = sl // self.mean_collapse_factor
            
        feature_dim = self._get_feature_dim(current_model.transformer.d_model, seq_len_after_collapse)

        # Create a classification head
        classifier_head = nn.Linear(feature_dim, num_classes).to(device, dtype=self.dtype)

        # Setup optimizer and loss
        # Combine parameters from the inverter, model, and the new head
        combined_params = list(current_inverter.parameters()) + list(current_model.parameters()) + list(classifier_head.parameters())
        optimizer = optim.AdamW(combined_params, lr=self.lr, weight_decay=self.weight_decay)
        criterion = nn.CrossEntropyLoss()

        # --- Training Loop ---
        current_model.train()
        current_inverter.train()
        classifier_head.train()
        
        for epoch in range(self.num_epochs):
            for batch_input, batch_label in train_dataloader:
                batch_input = batch_input.to(device, dtype=self.dtype)
                batch_label = batch_label.to(device, dtype=torch.long) # CrossEntropyLoss expects long labels

                # Reshape input
                window_size = batch_input.shape[2]
                batch_data = batch_input.reshape(self.batch_size, window_size // n_samples_per_bin, n_samples_per_bin)

                # Forward pass through inverter and model
                batch_data = current_inverter(batch_data)
                x_embed = current_model(batch_data) # Shape: (batch_size, seq_len, d_model)
                
                # Aggregate features
                features = self._aggregate_features(x_embed) # Shape: (batch_size, feature_dim)

                # Forward pass through classifier head
                logits = classifier_head(features) # Shape: (batch_size, num_classes)

                # Calculate loss
                loss = criterion(logits, batch_label)

                # Backward pass and optimization
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()

        # --- Evaluation Loop ---
        current_model.eval()
        current_inverter.eval()
        classifier_head.eval()

        all_probs = []
        all_labels = []
        
        with torch.no_grad():
            for batch_input, batch_label in test_dataloader:
                batch_input = batch_input.to(device, dtype=self.dtype)
                batch_label = batch_label.to(device, dtype=torch.long)

                window_size = batch_input.shape[2]
                batch_data = batch_input.reshape(self.batch_size, window_size // n_samples_per_bin, n_samples_per_bin)

                batch_data = current_inverter(batch_data)
                x_embed = current_model(batch_data)
                features = self._aggregate_features(x_embed)
                logits = classifier_head(features)
                
                probs = torch.softmax(logits, dim=1)
                
                all_probs.append(probs.cpu().numpy())
                all_labels.append(batch_label.cpu().numpy())

        all_probs = np.concatenate(all_probs)
        all_labels = np.concatenate(all_labels)

        # Calculate metrics
        accuracy = accuracy_score(all_labels, np.argmax(all_probs, axis=1))
        
        # Calculate AUROC
        if num_classes == 2:
             # Binary classification: use probabilities of the positive class (class 1)
             auroc = roc_auc_score(all_labels, all_probs[:, 1])
        else:
             # Multi-class classification
             try:
                 auroc = roc_auc_score(all_labels, all_probs, multi_class='ovr', average='macro')
             except ValueError as e:
                 print(f"Warning: Could not calculate AUROC ({e}). Setting AUROC to 0.0")
                 auroc = 0.0 # Handle cases where AUROC is undefined (e.g., only one class in test set)

        # Clean up GPU memory for the copied models and head for this split
        del current_model, current_inverter, classifier_head, optimizer, criterion
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
            
        return auroc, accuracy

    def evaluate(self):
        """Runs fine-tuning and evaluation across all specified tasks and splits."""
        results = {}
        # Iterate through unique subjects first
        unique_subjects = sorted(list(set(s.subject_identifier for s, _ in self.subject_trials)))
        
        for subject_id in unique_subjects:
            # Find the actual subject object (assuming identifier is unique)
            subject_obj = next(s for s, _ in self.subject_trials if s.subject_identifier == subject_id)
            
            for eval_name in self.eval_names:
                # Get trial IDs for this subject and eval_name
                trial_ids = [t_id for s, t_id in self.subject_trials if s.subject_identifier == subject_id]
                
                for trial_id in trial_ids:
                    dataset_key = (eval_name, subject_id, trial_id)
                    if dataset_key not in self.evaluation_datasets:
                        print(f"Warning: Splits not found for {dataset_key}. Skipping.")
                        continue
                        
                    splits = self.evaluation_datasets[dataset_key]
                    train_datasets, test_datasets = splits[0], splits[1] # Assuming this structure from generate_splits
                    
                    auroc_list, acc_list = [], []
                    print(f"Evaluating {subject_id} - Trial {trial_id} - Task {eval_name} ({len(train_datasets)} splits)")
                    
                    # Loop over cross-validation splits
                    for i, (train_dataset, test_dataset) in enumerate(zip(train_datasets, test_datasets)):
                        print(f"  Split {i+1}/{len(train_datasets)}...")
                        # Ensure datasets have enough samples for batching
                        if len(train_dataset) < self.batch_size or len(test_dataset) < self.batch_size:
                             print(f"  Warning: Split {i+1} has insufficient data (Train: {len(train_dataset)}, Test: {len(test_dataset)}). Skipping split.")
                             continue # Skip this split if too small

                        try:
                            auroc, acc = self.evaluate_on_dataset(train_dataset, test_dataset)
                            auroc_list.append(auroc)
                            acc_list.append(acc)
                            print(f"  Split {i+1} Results: AUROC={auroc:.4f}, Acc={acc:.4f}")
                        except Exception as e:
                            print(f"  Error during evaluation of split {i+1}: {e}")
                            # Optionally append NaN or skip failed splits
                            # auroc_list.append(np.nan)
                            # acc_list.append(np.nan)

                    if not auroc_list: # Handle case where all splits were skipped or failed
                         print(f"  No valid results for {dataset_key}.")
                         mean_auroc = np.nan
                         mean_acc = np.nan
                    else:
                        mean_auroc = np.nanmean(auroc_list) # Use nanmean in case of NaNs from failed splits
                        mean_acc = np.nanmean(acc_list)
                        print(f"Finished {dataset_key}: Mean AUROC={mean_auroc:.4f}, Mean Acc={mean_acc:.4f}")
                        
                    results[dataset_key] = (mean_auroc, mean_acc)
                    
        return self._format_results(results)
        
    def _format_results(self, results):
        """Formats the results dictionary for logging or display."""
        formatted_results = {}
        all_eval_aurocs = {eval_name: [] for eval_name in self.eval_names}
        all_eval_accs = {eval_name: [] for eval_name in self.eval_names}

        # Log individual trial results
        for (eval_name, subject_id, trial_id), (auroc, acc) in results.items():
            # Use nan_to_num to handle potential NaNs if needed, or keep them
            formatted_results[f"eval_auroc/{subject_id}_{trial_id}_{eval_name}"] = float(np.nan_to_num(auroc))
            formatted_results[f"eval_acc/{subject_id}_{trial_id}_{eval_name}"] = float(np.nan_to_num(acc))

        # Calculate average per subject per eval_name
        unique_subject_ids = sorted(list(set(sid for _, sid, _ in results.keys())))
        for eval_name in self.eval_names:
            subject_aurocs = {sid: [] for sid in unique_subject_ids}
            subject_accs = {sid: [] for sid in unique_subject_ids}
            
            for (en, sid, tid), (auroc, acc) in results.items():
                 if en == eval_name:
                     if not np.isnan(auroc): # Only include non-NaN results in averages
                         subject_aurocs[sid].append(auroc)
                     if not np.isnan(acc):
                         subject_accs[sid].append(acc)

            # Calculate mean over trials for each subject, then average subjects
            avg_subj_aurocs = []
            avg_subj_accs = []
            for sid in unique_subject_ids:
                 if subject_aurocs[sid]: # If list is not empty
                     avg_subj_aurocs.append(np.mean(subject_aurocs[sid]))
                 if subject_accs[sid]:
                     avg_subj_accs.append(np.mean(subject_accs[sid]))
            
            if avg_subj_aurocs: # If there are any valid subject averages
                 final_avg_auroc = np.mean(avg_subj_aurocs)
                 all_eval_aurocs[eval_name].append(final_avg_auroc) # Store for overall average
                 formatted_results[f"eval_auroc/average_{eval_name}"] = float(final_avg_auroc)
            else:
                 formatted_results[f"eval_auroc/average_{eval_name}"] = 0.0 # Or np.nan

            if avg_subj_accs:
                 final_avg_acc = np.mean(avg_subj_accs)
                 all_eval_accs[eval_name].append(final_avg_acc) # Store for overall average
                 formatted_results[f"eval_acc/average_{eval_name}"] = float(final_avg_acc)
            else:
                 formatted_results[f"eval_acc/average_{eval_name}"] = 0.0 # Or np.nan

        # Optional: Calculate overall average across all eval_names
        overall_aurocs = [v for name in self.eval_names for v in all_eval_aurocs[name]]
        overall_accs = [v for name in self.eval_names for v in all_eval_accs[name]]
        if overall_aurocs:
             formatted_results["eval_auroc/average_overall"] = float(np.mean(overall_aurocs))
        if overall_accs:
             formatted_results["eval_acc/average_overall"] = float(np.mean(overall_accs))
             
        return formatted_results
```

```python
import sklearn
from sklearn.linear_model import LogisticRegression
from torch.utils.data import DataLoader
from btbench.btbench_train_test_splits import generate_splits_SS_SM
import numpy as np

class ModelEvaluation():
    def __init__(self, model, inverter, subject_trials, eval_names, batch_size=128, dtype=torch.float32, feature_aggregation_method='mean', mean_collapse_factor=1):
        self.model = model
        self.inverter = inverter
        self.subject_trials = subject_trials
        self.eval_names = eval_names
        self.batch_size = batch_size
        self.dtype = dtype
        self.feature_aggregation_method = feature_aggregation_method
        self.mean_collapse_factor = mean_collapse_factor
        
        # Create evaluation datasets
        self.evaluation_datasets = {}
        for eval_name in self.eval_names:
            for subject, trial_id in self.subject_trials:
                splits = generate_splits_SS_SM(subject, trial_id, eval_name, dtype=self.dtype,
                                               start_neural_data_before_word_onset=0,
                                               end_neural_data_after_word_onset=2048)
                self.evaluation_datasets[(eval_name, subject.subject_identifier, trial_id)] = splits

    def evaluate_on_dataset(self, train_dataset, test_dataset):
        train_dataloader = DataLoader(train_dataset, batch_size=self.batch_size, shuffle=False, drop_last=True)
        test_dataloader = DataLoader(test_dataset, batch_size=self.batch_size, shuffle=False, drop_last=True)
        
        device = next(self.model.parameters()).device
        
        # Get embeddings for train and test data
        X_train, y_train = [], []
        for batch_input, batch_label in train_dataloader:
            batch_input = batch_input.to(device, dtype=self.dtype)
            window_size = batch_input.shape[2]
            batch_data = batch_input.to(device, dtype=dtype).reshape(self.batch_size, window_size//n_samples_per_bin, n_samples_per_bin) # shape (batch_size, seq_len, 1)
            batch_data = self.inverter(batch_data)

            with torch.no_grad():
                x_embed = self.model(batch_data)
            
            batch_size, seq_len, d_model = x_embed.shape
            x_embed = x_embed.reshape(self.batch_size, seq_len//self.mean_collapse_factor, self.mean_collapse_factor, d_model)
            x_embed = x_embed.mean(dim=2)

            if self.feature_aggregation_method == 'mean':
                X_train.append(x_embed.mean(dim=1).cpu().float().numpy())  # Average across sequence length
            elif self.feature_aggregation_method == 'concat':
                X_train.append(x_embed.reshape(self.batch_size, -1).cpu().float().numpy())  # Concatenate across sequence length
            y_train.append(batch_label.numpy())        
        X_test, y_test = [], []
        for batch_input, batch_label in test_dataloader:
            batch_input = batch_input.to(device, dtype=self.dtype)
            batch_data = batch_input.to(device, dtype=dtype).reshape(self.batch_size, window_size//n_samples_per_bin, n_samples_per_bin) # shape (batch_size, seq_len, 1)
            batch_data = self.inverter(batch_data)

            with torch.no_grad():
                x_embed = self.model(batch_data)
            batch_size, seq_len, d_model = x_embed.shape
            x_embed = x_embed.reshape(self.batch_size, seq_len//self.mean_collapse_factor, self.mean_collapse_factor, d_model)
            x_embed = x_embed.mean(dim=2)

            if self.feature_aggregation_method == 'mean':
                X_test.append(x_embed.mean(dim=1).cpu().float().numpy())
            elif self.feature_aggregation_method == 'concat':
                X_test.append(x_embed.reshape(self.batch_size, -1).cpu().float().numpy())
            y_test.append(batch_label.numpy())

        X_train = np.concatenate(X_train)
        y_train = np.concatenate(y_train)
        X_test = np.concatenate(X_test)
        y_test = np.concatenate(y_test)

        # Train logistic regression classifier
        clf = LogisticRegression(random_state=42, max_iter=10000)
        clf.fit(X_train, y_train)

        # Get predictions and calculate metrics
        test_probs = clf.predict_proba(X_test)
        accuracy = clf.score(X_test, y_test)
        
        # Calculate AUROC
        y_test_onehot = np.zeros((len(y_test), len(clf.classes_)))
        for i, label in enumerate(y_test):
            class_idx = np.where(clf.classes_ == label)[0][0]
            y_test_onehot[i, class_idx] = 1
            
        if len(clf.classes_) > 2:
            auroc = sklearn.metrics.roc_auc_score(y_test_onehot, test_probs, multi_class='ovr', average='macro')
        else:
            auroc = sklearn.metrics.roc_auc_score(y_test_onehot, test_probs)
            
        return auroc, accuracy

    def evaluate(self):
        results = {}
        for subject in set(subject for subject, _ in self.subject_trials):
            for eval_name in self.eval_names:
                trial_ids = [trial_id for _subject, trial_id in self.subject_trials if _subject == subject]
                for trial_id in trial_ids:
                    splits = self.evaluation_datasets[(eval_name, subject.subject_identifier, trial_id)]
                    auroc_list, acc_list = [], []
                    for train_dataset, test_dataset in zip(splits[0], splits[1]):
                        auroc, acc = self.evaluate_on_dataset(train_dataset, test_dataset)
                        auroc_list.append(auroc)
                        acc_list.append(acc)
                    
                    mean_auroc = np.mean(auroc_list)
                    mean_acc = np.mean(acc_list)
                    results[(eval_name, subject.subject_identifier, trial_id)] = (mean_auroc, mean_acc)
                    
        return self._format_results(results)
        
    def _format_results(self, results):
        formatted_results = {}
        for eval_name in self.eval_names:
            auroc_values = []
            acc_values = []
            subject_aurocs = {}
            subject_accs = {}
            
            for (metric, subject_id, trial_id) in [k for k in results.keys() if k[0] == eval_name]:
                if subject_id not in subject_aurocs:
                    subject_aurocs[subject_id] = []
                    subject_accs[subject_id] = []
                    
                auroc, acc = results[(eval_name, subject_id, trial_id)]
                subject_aurocs[subject_id].append(auroc)
                subject_accs[subject_id].append(acc)
                
                formatted_results[f"eval_auroc/{subject_id}_{trial_id}_{eval_name}"] = auroc
                formatted_results[f"eval_acc/{subject_id}_{trial_id}_{eval_name}"] = acc
                
            for subject_id in subject_aurocs:
                auroc_values.append(np.mean(subject_aurocs[subject_id]))
                acc_values.append(np.mean(subject_accs[subject_id]))
                
            if auroc_values:
                formatted_results[f"eval_auroc/average_{eval_name}"] = np.mean(auroc_values)
                formatted_results[f"eval_acc/average_{eval_name}"] = np.mean(acc_values)
                
        return formatted_results
```

## 4. Training the model

```python
initial_lr = 0.003
use_muon = True
optimizers = []
schedulers = []
if use_muon:
    from muon import Muon
    all_params = list(model.parameters())
    matrix_params = [p for p in all_params if p.ndim >= 2]
    other_params = [p for p in all_params if p.ndim < 2]
    optimizers.append(Muon(matrix_params, lr=initial_lr, momentum=0.95, nesterov=True, backend='newtonschulz5', backend_steps=5))
    if len(other_params) > 0:
        optimizers.append(torch.optim.AdamW(other_params, lr=initial_lr, betas=(0.9, 0.95)))
    #schedulers.append(None)  # Muon doesn't support schedulers
    #schedulers.append(torch.optim.lr_scheduler.LinearLR(optimizers[1], start_factor=1.0, end_factor=0.0, total_iters=n_steps))
else:
    optimizers = [torch.optim.AdamW(model.parameters(), lr=initial_lr, betas=(0.9, 0.95))]
    #schedulers = [torch.optim.lr_scheduler.LinearLR(optimizers[0], start_factor=1.0, end_factor=0.0, total_iters=n_steps)]

training_logs = []
step = 1
for batch in dataloader:
    for optimizer in optimizers:
        optimizer.zero_grad()

    batch_data = batch['data'].to(device, dtype=dtype).reshape(batch_size, window_size//n_samples_per_bin, n_samples_per_bin) # shape (batch_size, seq_len, 1)
    batch_data = inverter(batch_data)

    loss = model.calculate_loss(batch_data)
    loss.backward()
    for optimizer in optimizers:
        optimizer.step()
    
    # Step the schedulers
    for scheduler in schedulers:
        if scheduler is not None:
            scheduler.step()
    
    # Log metrics
    log_dict = {
        'train_loss': loss.item(),
        'unembed_std': model.embed.embedding_weights.data.cpu().float().numpy().std()
    }
    
    if step % 100 == 0:
        current_lr = optimizers[-1].param_groups[0]['lr']
        print(f"Step {step}, Loss: {loss.item():.4f}, LR: {current_lr:.6f}")
        print("\t std of embedding weights: ", log_dict['unembed_std'])
        
        # Add evaluation results
        evaluation_results = evaluation.evaluate()
        log_dict.update(evaluation_results)
        print(log_dict)
        
    training_logs.append(log_dict)
        
    if step == n_steps:
        break # Only process one batch per step
    step += 1
```
