```python
from subject_mgh import MGHSubject
import pandas as pd
```

```python
subject = MGHSubject(1)
# Get all electrode labels across sessions
all_electrode_labels = subject.get_electrode_labels()

# Count how many sessions each electrode appears in
electrode_counts = {electrode: [] for electrode in all_electrode_labels}
for electrode in all_electrode_labels:
    for session_id in range(len(subject.sessions)):
        if electrode in subject.get_electrode_labels(session_id):
            electrode_counts[electrode].append(session_id)
    

# Extract probe names from electrode labels
probe_counts = {}
for electrode in all_electrode_labels:
    # Get probe name by removing trailing numbers
    probe_name = ''.join(c for c in electrode if not c.isdigit())
    probe_counts[probe_name] = probe_counts.get(probe_name, []) + electrode_counts[electrode]
#electrode_counts = probe_counts

EEG_electrode_labels = ['Fp1', 'Fp2', 'F3', 'Fz', 'F4', 'C3', 'Cz', 'C4', 'P3', 'Pz', 'P4', 'O1', 'O2', 'F7', 'F8', 'T3', 'T4', 'T5', 'T6']
EEG_electrode_labels = [e.upper() for e in EEG_electrode_labels]
electrode_counts = {k: v for k, v in electrode_counts.items() if k.upper() in EEG_electrode_labels}


# Sort electrodes by count
sorted_electrodes = sorted(electrode_counts.items(), key=lambda x: x[1], reverse=True)
sorted_electrode_labels = [e[0] for e in sorted_electrodes]

n_cols = 1
for i in range(0, len(sorted_electrode_labels), n_cols):
    print("\t\t\t".join(sorted_electrode_labels[i:i+n_cols]))
    print("\t\t\t".join([str(electrode_counts[e]) for e in sorted_electrode_labels[i:i+n_cols]]))
    print()
```
