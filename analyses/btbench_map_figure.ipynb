```python
# from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.pyplot as plt

matlab_xlim = (-108.0278, 108.0278)
matlab_ylim = (-72.9774, 72.9774)

base_path = os.path.join(ROOT_DIR, 'localization')
left_hem_file_name = 'left_hem_clean.png'
right_hem_file_name = 'right_hem_clean.png'
coords_file_name = 'elec_coords_full.csv'
correlations_file = 'lag_correlation.json'

left_hem_img = plt.imread(os.path.join(base_path, left_hem_file_name))
right_hem_img = plt.imread(os.path.join(base_path, right_hem_file_name))
coords_df = pd.read_csv(os.path.join(base_path, coords_file_name))
split_elec_id = coords_df['ID'].str.split('-')
coords_df['Subject'] = [t[0] for t in split_elec_id]
coords_df['Electrode'] = [t[1] for t in split_elec_id]

# Scale Matlab electrode locations to Python format
def scale(x, s, d):
    return -(x - d) * s

x_scale = left_hem_img.shape[1] / (matlab_xlim[1] - matlab_xlim[0])
y_scale_l = left_hem_img.shape[0] / (matlab_ylim[1] - matlab_ylim[0])

y_scale_r = right_hem_img.shape[0] / (matlab_ylim[1] - matlab_ylim[0])

scaled_coords_df = coords_df.copy()

# scale left hemisphere coordinates
scaled_coords_df.loc[scaled_coords_df['Hemisphere'] == 1, 'X'] = coords_df.loc[coords_df['Hemisphere'] == 1, 'X'].apply(lambda x: scale(x, x_scale, matlab_xlim[1]))
scaled_coords_df.loc[scaled_coords_df['Hemisphere'] == 1, 'Y'] = coords_df.loc[coords_df['Hemisphere'] == 1, 'Y'].apply(lambda x: scale(x, y_scale_l, matlab_ylim[1]))

# scale right hemisphere coordinates
scaled_coords_df.loc[scaled_coords_df['Hemisphere'] == 0, 'X'] = coords_df.loc[coords_df['Hemisphere'] == 0, 'X'].apply(lambda x: -scale(x, y_scale_r, matlab_xlim[0]))
scaled_coords_df.loc[scaled_coords_df['Hemisphere'] == 0, 'Y'] = coords_df.loc[coords_df['Hemisphere'] == 0, 'Y'].apply(lambda x: scale(x, y_scale_r, matlab_ylim[1]))

def plot_hemisphere_axis(electrodes, colors=None, ax=None, hemisphere="left", title=None, vmin=0, vmax=0.1, inverse_order=False):
    '''
    electrodes is a dict of format {<subject>: [<electrode>]}
    '''
    if colors is None: c = 'white'
    ax.set_aspect('equal')

    if hemisphere=="left":
        ax.imshow(left_hem_img)
    elif hemisphere=="right":
        ax.imshow(right_hem_img)
    
    ax.axis('off')
    assert hemisphere in ["left", "right"]
    hem_index = 1 if hemisphere=="left" else 0

    plot_title = f'{hemisphere} hemisphere'
    if title: plot_title += f' {title}'
   
    all_x, all_y, all_colors = [], [], []
    for s in electrodes:
        for e in electrodes[s]:
            x = list(scaled_coords_df[(scaled_coords_df.Subject == s) & (scaled_coords_df.Electrode == e) & (scaled_coords_df.Hemisphere==hem_index)]['X'])
            y = list(scaled_coords_df[(scaled_coords_df.Subject == s) & (scaled_coords_df.Electrode == e) & (scaled_coords_df.Hemisphere==hem_index)]['Y'])

            all_x += x
            all_y += y
            all_colors += [colors[e]] * len(x)
    
    all_x = np.array(all_x)
    all_y = np.array(all_y)
    all_colors = np.array(all_colors)
    # Sort points by color value so higher values appear on top
    # sort_idx = np.argsort(all_colors)[::-1] if inverse_order else np.argsort(all_colors)
    # all_x = all_x[sort_idx]
    # all_y = all_y[sort_idx]
    # all_colors = all_colors[sort_idx]
    cmap = 'viridis_r' if inverse_order else 'viridis'
    sc = ax.scatter(all_x, all_y, c=all_colors, vmin=vmin, vmax=vmax, s=100, edgecolors='white', cmap=cmap)
    
def plot_hemispheres_separately(electrodes, ax1, ax2, colors=None, vmin=0, vmax=0.1, inverse_order=False):
    plot_hemisphere_axis(electrodes, colors=colors, ax=ax1, hemisphere="right", vmin=vmin, vmax=vmax, inverse_order=inverse_order)
    plot_hemisphere_axis(electrodes, colors=colors, ax=ax2, hemisphere="left", vmin=vmin, vmax=vmax, inverse_order=inverse_order)
```

```python
def load_results(subject_id, trial_id):
    filename = f'analyses/eval_results/frozen_transformer_model_electrode_time_bins_btbank{subject_id}_{trial_id}.json'
    with open(filename) as f:
        return json.load(f)

def get_mean_auroc(results, electrode, time_bin_idx):
    """Get mean AUROC across folds for an electrode at a specific time bin"""
    electrode_data = results['evaluation_results'][f'btbank{results["subject"]}_{results["trial"]}']['electrode'][electrode]
    time_bin = electrode_data['time_bins'][time_bin_idx]
    aurocs = [fold['test_roc_auc'] for fold in time_bin['folds']]
    return np.mean(aurocs)

def get_pvalue(results, electrode, time_bin_idx):
    """Get p-value from one-sided t-test comparing AUROC to 0.5"""
    electrode_data = results['evaluation_results'][f'btbank{results["subject"]}_{results["trial"]}']['electrode'][electrode]
    time_bin = electrode_data['time_bins'][time_bin_idx]
    aurocs = [fold['test_roc_auc'] for fold in time_bin['folds']]
    from scipy import stats
    t_stat, p_val = stats.ttest_1samp(aurocs, 0.5)
    # One-sided p-value (testing if mean > 0.5)
    p_val = p_val / 2 if t_stat > 0 else 1 - p_val / 2
    return p_val

def plot_electrodes_for_timebin(time_bin_idx, plot_pvalue=False, vmin=None, vmax=None):
    # Load results for each subject/trial pair
    subject_trials = [(10,0), (1,2), (2,6), (3,0), (7,0)]
    all_electrodes = {}
    colors = {}
    
    # Set default vmin/vmax based on metric
    if vmin is None:
        vmin = 0 if plot_pvalue else 0.5
    if vmax is None:
        vmax = 0.05 if plot_pvalue else 0.6
    
    for subject_id, trial_id in subject_trials:
        results = load_results(subject_id, trial_id)
        results['subject'] = subject_id
        results['trial'] = trial_id
        
        # Get electrode labels and their values
        electrodes = list(results['evaluation_results'][f'btbank{subject_id}_{trial_id}']['electrode'].keys())
        all_electrodes[f'sub_{subject_id}'] = electrodes
        
        for electrode in electrodes:
            if plot_pvalue:
                value = get_pvalue(results, electrode, time_bin_idx)
            else:
                value = get_mean_auroc(results, electrode, time_bin_idx)
            colors[electrode] = value

    # Create plot
    fig, axes = plt.subplots(1, 2, figsize=(14, 5),
                            gridspec_kw={'wspace': 0.1})
    
    # Plot hemispheres
    cmap = 'viridis_r' if plot_pvalue else 'viridis'
    plot_hemispheres_separately(all_electrodes, axes[0], axes[1], colors=colors, vmin=vmin, vmax=vmax, inverse_order=plot_pvalue)
    
    # Add colorbar
    norm = plt.Normalize(vmin=vmin, vmax=vmax)
    sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
    colorbar_label = 'p-value (one-sided t-test)' if plot_pvalue else 'Mean decoding AUROC'
    plt.colorbar(sm, ax=axes, label=colorbar_label)
    
    time_bin_start = results['evaluation_results'][f'btbank{subject_id}_{trial_id}']['electrode'][electrodes[0]]['time_bins'][time_bin_idx]['time_bin_start']
    time_bin_end = results['evaluation_results'][f'btbank{subject_id}_{trial_id}']['electrode'][electrodes[0]]['time_bins'][time_bin_idx]['time_bin_end']
    metric_name = 'p-value' if plot_pvalue else 'Mean AUROC'
    plt.suptitle(f'Task: Sentence Onset\n{metric_name} by Electrode (Time: {time_bin_start}s to {time_bin_end}s)')
    
    os.makedirs('analyses/btbench_map_figure/', exist_ok=True)
    metric_suffix = 'pvalue' if plot_pvalue else 'auroc'
    plt.savefig(f'analyses/btbench_map_figure/hemispheres_electrodes_{metric_suffix}_timebin{time_bin_idx}.png', bbox_inches='tight', dpi=300, transparent=False)
    plt.close()
```

```python
num_time_bins = 24

# Generate plots for all time bins
for time_bin_idx in range(num_time_bins):
    plot_electrodes_for_timebin(time_bin_idx, plot_pvalue=True)
    plot_electrodes_for_timebin(time_bin_idx, plot_pvalue=False)
    print(f"Generated plot for time bin {time_bin_idx}")
```

# Plots over time

```python
subject_id, trial_id = 3, 0
results = load_results(subject_id, trial_id)['evaluation_results'][f'btbank{subject_id}_{trial_id}']['electrode']
```

```python
electrode_labels = list(results.keys())

show_n_electrodes = 10
import random
electrode_labels_selected = random.sample(electrode_labels, show_n_electrodes)

for electrode in electrode_labels_selected:
    # Extract time bins data
    time_bins = results[electrode]['time_bins']
    
    # Prepare data for plotting
    time_points = [(tb['time_bin_start'] + tb['time_bin_end'])/2 for tb in time_bins]
    test_accuracies = []
    test_accuracies_sem = []
    
    # Calculate mean and SEM for each time bin
    for time_bin in time_bins:
        fold_accuracies = [fold['test_accuracy'] for fold in time_bin['folds']]
        mean_acc = np.mean(fold_accuracies)
        sem_acc = np.std(fold_accuracies) / np.sqrt(len(fold_accuracies))
        test_accuracies.append(mean_acc)
        test_accuracies_sem.append(sem_acc)
    
    # Plot with error bars
    plt.figure(figsize=(5, 3))
    plt.errorbar(time_points, test_accuracies, yerr=test_accuracies_sem, 
                label=electrode, marker='o', capsize=3)
    
    plt.xlabel('Time (s)')
    plt.ylabel('Test Accuracy')
    plt.title(f'Decoding Accuracy Over Time\nElectrode: {electrode}')
    plt.grid(True)
    plt.axhline(y=0.5, color='r', linestyle='--', alpha=0.5)  # Chance level
    plt.legend()
    plt.tight_layout()
    plt.show()
```
