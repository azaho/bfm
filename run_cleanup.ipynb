```python
import subprocess

# Run the command and capture output
cmd = "sinfo -N | awk '{print $1}' | grep -v NODELIST"
result = subprocess.run(cmd, shell=True, capture_output=True, text=True)

# Split output into lines and store in array
node_list = result.stdout.strip().split('\n')
node_list = list(set(node_list))
```

```python
processes = []
for node in node_list[:]:
    # Create temporary sbatch script
    script_content = f"""#!/bin/bash
#SBATCH --nodelist={node}
#SBATCH --ntasks-per-node=1
#SBATCH --mem-per-cpu=100M
#SBATCH --time=1:00:00
#SBATCH --output=logs/{node}.out

echo "Size before cleanup:"
du -sh $(find /dev/shm -user $(whoami) -type f 2>/dev/null | tr "\\n" " ") 2>/dev/null || echo "0"
echo "Running cleanup..."
find /dev/shm -user $(whoami) -delete
echo "Size after cleanup:" 
du -sh $(find /dev/shm -user $(whoami) -type f 2>/dev/null | tr "\\n" " ") 2>/dev/null || echo "0"
"""
    
    # Write script to temporary file
    script_file = f"logs/cleanup_{node}.sh"
    with open(script_file, "w") as f:
        f.write(script_content)
    
    # Submit job
    print(node)
    cmd = f"sbatch {script_file}"
    p = subprocess.Popen(cmd, shell=True)
    p.wait()

# Clean up temporary script files
for node in node_list:
    script_file = f"logs/cleanup_{node}.sh"
    subprocess.run(f"rm {script_file}", shell=True)
```

```python
processes = []
for node in node_list:
    cmd = f"""srun --nodelist={node} --ntasks-per-node=1 --mem-per-cpu=100M --time=1:00:00 --mpi=none bash -c '
        echo "Size before cleanup:"
        du -sh $(find /dev/shm -user $(whoami) -type f 2>/dev/null | tr "\\n" " ") 2>/dev/null || echo "0"
        echo "Running cleanup..."
        find /dev/shm -user $(whoami) -delete
        echo "Size after cleanup:"
        du -sh $(find /dev/shm -user $(whoami) -type f 2>/dev/null | tr "\\n" " ") 2>/dev/null || echo "0"
    ' > logs/{node}.out 2>&1"""
    print(node)
    p = subprocess.Popen(cmd, shell=True)
    processes.append(p)

# Wait for all processes to complete
for p in processes:
    p.wait()
```
